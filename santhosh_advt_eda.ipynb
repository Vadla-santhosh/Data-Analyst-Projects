```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from warnings import filterwarnings
filterwarnings("ignore")
```

```python
df = pd.read_csv("Downloads/advertising.csv")
df
```

```python
df.info()
```

```python
df.describe(include='all').T
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
plt.plot(df["Age"],color="r")
plt.title('Age')
plt.figure(figsize=(500,500))
plt.show()
```

```python
plt.plot(df["Area Income"],color="b")
plt.title('Area of income')
plt.figure(figsize=(500,500))
plt.show()
```

```python
plt.barh(df["Age"],df["Clicked on Ad"],color="k")
plt.title('horizontal age by clicked on ad')
plt.figure(figsize=(2000,1000))
plt.show()
```

```python
plt.bar(df["Age"],df["Daily Time Spent on Site"],color="c")
plt.title('Age by daily time spent')
plt.figure(figsize=(2000,1000))
plt.show()
```

```python
d=df['Ad Topic Line']
g=df['Country']
plt.figure(figsize = (10,5))
plt.scatter(d[0:20], g[0:20], color="orange")
plt.title('Ad topic line by country')
plt.xticks(rotation = 90)
plt.show()
```

```python
d=df['Ad Topic Line']
g=df['Country']
plt.figure(figsize = (10,5))
plt.scatter(d[0:30], g[0:30], color="orange")
plt.title('scatter/plot Ad topic line by country')
plt.plot(d[0:30], g[0:30], color="k")
plt.xticks(rotation = 90)
plt.show()
```

```python
v=df['Timestamp']
u=df['Age']
plt.figure(figsize=(20,8))
plt.scatter(v[0:40], u[0:40], c=u[0:40], linewidth=10, cmap ='nipy_spectral')
plt.title('Time stamp by age')
plt.colorbar()
plt.xticks(rotation=90)
plt.show()
```

```python
v=df['Timestamp']
u=df['Age']
plt.figure(figsize=(20,8))
plt.scatter(v[0:40], u[0:40], c=u[0:40], linewidth=10, cmap = plt.cm.plasma)
plt.title('Time stamp by Age')
plt.colorbar()
plt.xticks(rotation=90)
plt.show()
```

```python
plt.pie(df['Daily Internet Usage'])
plt.show()
```

```python
plt.title('histogram:Count on Daily time spent on site')
sns.histplot(df['Daily Time Spent on Site'])
```

```python
plt.title('Count of Area income')
sns.histplot(df['Area Income'], bins=50, color='maroon')
```

```python
plt.title('Count of Daily internet usage')
sns.histplot(df['Daily Internet Usage'], kde=True, color='yellow');
```

```python
sns.pairplot(df, hue='Age')
```

```python
sns.jointplot(x=df['Area Income'], y=df['Age'], kind='hex', color='r')
```

```python
sns.jointplot(x=df['Age'], y=df['Daily Time Spent on Site'], kind='kde', fill=True, color='b')
```

```python
sns.distplot(df['Daily Internet Usage'], color='g')
plt.title('Daily internet usage')
```

```python
plt.figure(figsize = (30,8))
sns.relplot(x=df['Age'], y=df['Area Income'], color='purple')
plt.title('Age by income')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize = (10,5))
sns.boxplot(df, x="Age", y="Area Income")
plt.title('Age by area of income')
plt.xticks(rotation=90)
plt.show()
```

```python
g=df['Daily Time Spent on Site']
h=df['Area Income']
plt.figure(figsize = (10,5))
sns.violinplot(x=h[0:30], y=g[0:30])
plt.title('Daily time spent by area of income')
plt.xticks(rotation=90)
plt.show()
```
