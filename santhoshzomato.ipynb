```python
plt.title('Number of Votes')
plt.plot(d['Votes'], color='red')
plt.xlabel('votes')
plt.ylabel('count')
plt.show()
```

```python
plt.title('Average Cost for two')
plt.plot(d['Average Cost for two'], color='blue')
plt.show()
```

```python
plt.figure(figsize = (15,10))
plt.title('Restaurant name with prices')
x=d["Restaurant Name"]
y=d["Price range"]
plt.xlabel("Restaurant Name")
plt.ylabel("Price range")
plt.bar(x[0:20],y[0:20], color="orange")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.title('count of price range')
sns.countplot(x=d['Price range'])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.title('histogram:Latitude and longitude')
sns.histplot(data=d, x='Latitude', y='Longitude', bins=50)
```

```python
plt.title('histogram:Aggregate rating')
sns.histplot(data=d, x='Aggregate rating', bins=50, color='orange')
```

```python
plt.title('histogram kde: Price range')
sns.histplot(data=d, x='Rating color', hue='Rating color', kde=True)
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.title('Distplot:Density')
sns.distplot(d['Restaurant ID'], color='blue')
```

```python
plt.figure(figsize = (20,10))
g = sns.FacetGrid(d, col="Price range")
g.map(sns.histplot, "Price range", color='k') 
plt.show()
```

```python
plt.title('Restaurant id with price range')
plt.scatter(d['Price range'], d['Restaurant ID'], alpha=0.2, color='violet')
plt.xlabel('Price range')
plt.ylabel('Restaurant ID')
plt.show()
```

```python
plt.figure(figsize = (25,10))
plt.scatter(d['City'], d['Price range'], alpha = 0.5, color='Orange')
plt.title('City wise price ranges')
plt.xlabel('City')
plt.ylabel('Price range')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.title('votes according to rating color')
plt.scatter(d['Rating color'], d['Votes'], c=d['Votes'], alpha = 0.5, cmap = 'nipy_spectral')
plt.xlabel('Rating color')
plt.ylabel('Votes')
plt.show()
```

```python
sns.pairplot(d, hue='Price range')
```

```python
sns.jointplot(x=d['Aggregate rating'], y=d['Restaurant ID'], kind='hex', color='magenta')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.jointplot(x=d['Average Cost for two'], y=d['Price range'], kind='reg', color='k')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize = (10,10))
sns.boxplot(data=d, x="Price range", y="Rating color")
plt.title('Price range wise rating color')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize = (10,5))
sns.violinplot(y=d['Aggregate rating'], x=d['Rating text'])
plt.title('Aggregate rating wise rating text')
plt.xticks(rotation=90)
plt.show()
```
