```python
plt.title('Age of interest')
plt.bar(data['age'], data['interest'], color='red')
plt.xlabel('age')
plt.ylabel('interest')
plt.xticks(rotation = 90)
plt.yticks(rotation = 45)
plt.show()
```

```python
plt.figure(figsize = (20,10))
sns.countplot(x=data['interest'])
plt.title('Count of Interest')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.title('histogram:Sugars', color='red')
sns.histplot(data=data, x='Clicks', bins=50, color='yellow')
```

```python
plt.title('histogram kde: Genderwise Interest ')
sns.histplot(data=data, x='interest', hue='gender', kde=True)
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.distplot(data['Approved_Conversion'], color='c')
plt.title('Approved conversion')
```

```python
g = sns.FacetGrid(data, col="age")
g.map(sns.histplot, "Clicks", color='red')
```

```python
plt.pie(x=data['Impressions'])
plt.legend({'Male', 'Female'})
plt.show()
```

```python
plt.title('sctter and plot of Interest wise approved conversion')
plt.scatter(data['interest'], data['Approved_Conversion'], color='red')
plt.plot(data['interest'], (data['Approved_Conversion']), color = 'black')
plt.colorbar()
plt.show()
```

```python
plt.title('Interest wise approved_conversion')
plt.scatter(data['interest'], data['Approved_Conversion'], alpha = 0.5, color='indigo')
plt.xlabel('Interest')
plt.ylabel('Approved_Conversion')
plt.colorbar()
plt.show()
```

```python
plt.title('Approved_conversions by Total_conversion')
plt.scatter(data['Approved_Conversion'], data['Total_Conversion'], alpha = 0.5, color='Orange')
plt.xlabel('Approved_Conversion')
plt.ylabel('Total_Conversion')
plt.show()
```

```python
plt.title('Amount spent on interest')
plt.scatter(data['Spent'], data['interest'], c=data['interest'], alpha = 0.5, cmap = 'nipy_spectral')
plt.xlabel('Spent')
plt.ylabel('interest')
plt.show()
```

```python
plt.figure(figsize = (10,5))
sns.pairplot(data, hue='age')
```

```python
sns.jointplot(x=data['fb_campaign_id'], y=data['Spent'], kind='hex', color='green')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.jointplot(x=data['xyz_campaign_id'], y=data['Spent'], kind='reg', color='orange')
```

```python
plt.figure(figsize = (10,5))
sns.boxplot(data=data, x="age", y="Clicks")
plt.title('Age wise no. ofclicks')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize = (10,5))
sns.violinplot(y=data['Total_Conversion'], x=data['age'])
plt.title('Total_conversion with age')
plt.xticks(rotation=90)
plt.show()
```
